#1. Apply Beautifulsoup algorithm and extract information about the corona virus from any sites available.

```
from bs4 import BeautifulSoup
import requests
```

```
url="https://kidshealth.org/en/parents/coronavirus.html"
```

```
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify()) # print the parsed data of html
```

```
print(soup.title.text)
```

```
text = soup.find_all(text=True)
```

```
set([t.parent.name for t in text])
```

```
output = ''
blacklist = [
    '[document]',
    'noscript',
    'header',
    'html',
    'meta',
    'head', 
    'input',
    'script',
    ]

for t in text:
    if t.parent.name not in blacklist:
        output += '{} '.format(t)
print(output)
```

# 2. Print the top 10 important words

```
from nltk.tokenize import word_tokenize,sent_tokenize
nltk_tokens=nltk.sent_tokenize(output)
print(nltk_tokens)
```

```
nltk_word_tokens = []
for i in nltk_tokens:
    nltk_word_tokens.append(nltk.word_tokenize(i))
for j in nltk_word_tokens:
    print(j)    
```

```
nltk_word_tokens = nltk.word_tokenize(output)
nltk_word_tokens=[nltk_word_token.lower() for nltk_word_token in nltk_word_tokens if nltk_word_token.isalpha()]
frequency = nltk.FreqDist(word.lower() for word in nltk_word_tokens)
```

```
for nltk_word_token, freq in frequency.most_common(10):
    print('%s ,%d' % (nltk_word_token, freq)) 
```
